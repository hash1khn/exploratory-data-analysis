```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
# load data from HW4.csv into data frame
df=pd.read_csv('/content/HW4.csv',header=0)
# report the data frame
df
```

```python
# EDA:
# print data frame info
print("Data Frame Info:\n")
print(df.info())

# report all descriptive statistics
print("Data Frame Description")
print(df.describe())
```

```python
# report correlation matrix in color map
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# feature 0:
# print all unique values in the ascending order
feature_0 = df['25']
unique_values = feature_0.unique()
sorted_unique_values = sorted(unique_values)
print("All Unique Values:")
print(sorted_unique_values)

# report each value and count in the ascending order on values
value_counts_sorted = pd.Series(sorted_unique_values).value_counts().sort_index()
for value, count in value_counts_sorted.items():
    print(f"{value}: {count}")
```

```python
# report records with largest feature 1
largest_index = df['1'].idxmax()
print(largest_index)
```

```python
# print count of records above the average feature 5
average_feature_5 = df['49'].mean()
count_above_average = (df['49'] > average_feature_5).sum()

print("Count of records above the average feature 5:", count_above_average)
```

```python
# report records, in which features 8, 9, 10, 11 and 12 are all 1
filtered_records = df[(df['0.1'] == 1) & (df['1.2'] == 1) & (df['0.2'] == 1) & (df['0.3'] == 1) & (df['0.4'] == 1)]

# Print the filtered records
print(filtered_records)
```

```python
# for each value feature 2, what are the averages per other feature
averages_per_feature2 = df.groupby('1.1').mean()

# Print the averages per other feature for each value of feature 2
print(averages_per_feature2)
```

```python
# for each combination among five features 8, 9, 10, 11 and 12, how many records
# per combination (report results using only feature 0)
combination_counts = df.groupby(['1.60', '0.1', '1.2', '0.2', '0.3'])['25'].count()

# Print the counts for each combination
print(combination_counts)
```

```python
# report frequency counts on feature 3
frequency_counts = df['4'].value_counts()
# label horizontal axis as Feature 3 and show values only in integers
frequency_counts.plot(kind='bar')

# Set labels and format axes
plt.xlabel('Feature 3')
plt.ylabel('Count')# label vertical axis as Count
plt.xticks(rotation=0)  # Rotate x-axis labels to horizontal
plt.yticks(frequency_counts.values)  # Show y-axis values only in integers

# force to show the plot
plt.show()
```

```python
# report three histogram subplots on one 15"x3" figure as follows:

# on the left, report distribution on feature 5 with proper labels

# on the center, report distribution on feature 6 with proper labels

# on the right, report distribution on feature 7 with proper labels

fig, axs = plt.subplots(1, 3, figsize=(15, 3))

# Plot histogram for feature 5
axs[0].hist(df['49'], bins=10, color='skyblue', edgecolor='black')
axs[0].set_title('Feature 5')
axs[0].set_xlabel('Feature 5 Values')
axs[0].set_ylabel('Frequency')

# Plot histogram for feature 6
axs[1].hist(df['0'], bins=10, color='salmon', edgecolor='black')
axs[1].set_title('Feature 6')
axs[1].set_xlabel('Feature 6 Values')
axs[1].set_ylabel('Frequency')

# Plot histogram for feature 7
axs[2].hist(df['1.60'], bins=10, color='lightgreen', edgecolor='black')
axs[2].set_title('Feature 7')
axs[2].set_xlabel('Feature 7 Values')
axs[2].set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the figure
plt.show()
```

```python
# report five pie subplots on one 15"x3" figure as follows:

# from left to right, report distribution on features 8, 9, 10, 11 and 12, each
# pie should be labeled with 0 and 1 with percentage on the two same colors

colors = ['skyblue', 'salmon']

# Create subplots
fig, axs = plt.subplots(1, 5, figsize=(15, 3))

# Plot each pie chart individually
axs[0].pie(df['0.1'].value_counts(normalize=True), labels=['1', '0'], autopct='%1.1f%%', colors=colors)
axs[0].set_title('Feature 0.1')
axs[0].axis('equal')

axs[1].pie(df['1.2'].value_counts(normalize=True), labels=['1', '0'], autopct='%1.1f%%', colors=colors)
axs[1].set_title('Feature 1.2')
axs[1].axis('equal')

axs[2].pie(df['0.2'].value_counts(normalize=True), labels=['1', '0'], autopct='%1.1f%%', colors=colors)
axs[2].set_title('Feature 0.2')
axs[2].axis('equal')

axs[3].pie(df['0.3'].value_counts(normalize=True), labels=['1', '0'], autopct='%1.1f%%', colors=colors)
axs[3].set_title('Feature 0.3')
axs[3].axis('equal')

axs[4].pie(df['0.4'].value_counts(normalize=True), labels=['1', '0'], autopct='%1.1f%%', colors=colors)
axs[4].set_title('Feature 0.4')
axs[4].axis('equal')

# Adjust layout to prevent overlap
plt.tight_layout()

# Force to show the figure
plt.show()
```

```python
# report count of feature 2 for each value of feature 4 using a treemap
# Group by feature 4 and feature 2, and count the occurrences
grouped_data = df.groupby(['91107', '1.1']).size().reset_index(name='count')

# Convert '91107' column to string
grouped_data['91107'] = grouped_data['91107'].astype(str)

# Create a treemap
fig = px.treemap(grouped_data, path=['91107', '1.1'], values='count')
fig.update_layout(title_text='Treemap of feature 2 count for each value of feature 4')
# Show the plot
fig.show()
```
